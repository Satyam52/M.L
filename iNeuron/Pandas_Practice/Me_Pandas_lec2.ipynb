### Getting Data Online

```python
import requests
url = 'https://api.github.com/repos/pandas-dev/pandas/issues'
res = requests.get(url)
res
```

```python
data = res.json()
data[2]
```

```python
df = pd.DataFrame(data,columns=['url','comments_url','html_url','user','labels'])
```

```python
d = dict()
d['url'] = df['url']
d['comments_url'] = df['comments_url']
d['html_url'] = df['html_url']
```

```python
# For user
for i in range(len(df['user'])):  # i = index
    for j in df['user'][i]:       # j = key
        if i == 0:
            d[j] = []
        d[j].append(df['user'][i][j])
```

```python
# For Labels
for i in range(len(df['labels'])):  # i = index
    for j in range(len(df['labels'][i])):   
        for k in df['labels'][i][j]:
            if i == 0:
                d[k] = []
            d[k].append(df['labels'][i][j][k])
```

```python
xyz = pd.DataFrame(d)
# d
```
